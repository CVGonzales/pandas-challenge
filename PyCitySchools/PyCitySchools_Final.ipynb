```python
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
schools_count = len(school_data_complete["school_name"].unique())
print(schools_count)

student_count = school_data_complete["student_name"].count()
print(student_count)

total_budget = sum(school_data_complete["budget"].unique())
print('${:,.2f}'.format(total_budget))

avg_math = school_data_complete["maths_score"].mean()
print(avg_math)

avg_read = school_data_complete["reading_score"].mean()
print(avg_read)

math_pass = school_data_complete.loc[school_data_complete["maths_score"] >= 50]
math_pass_per = len(math_pass)/ student_count * 100
print(math_pass_per)

read_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 50]
read_pass_per = len(read_pass)/ student_count * 100
print(read_pass_per)

both_pass = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]
both_pass_per = len(both_pass)/ student_count * 100
print(both_pass_per)
```

```python
LGA_summary = pd.DataFrame({"Total Schools": schools_count,
                           "Total Students": student_count,
                           "Total Budget": "${:,.2f}".format(total_budget),
                           "Average Maths Score": avg_math,
                           "Average Reading Score": avg_read,
                           "% Passing Maths": math_pass_per,
                           "% Passing Reading": read_pass_per,
                           "% Overall Passing Rate": [both_pass_per]})
LGA_summary
```

```python
grouped_school = school_data_complete.groupby(["school_name"])

school_type = grouped_school["type"].first()
school_students = grouped_school["size"].first()
school_budget = grouped_school["budget"].first()
student_budget = school_budget / school_students
school_avg_math = grouped_school["maths_score"].mean()
school_avg_read = grouped_school["reading_score"].mean()

school_math_pass = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["school_name"])["maths_score"].count()
school_math_pass_per = school_math_pass/ school_students * 100

school_read_pass = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["school_name"])["reading_score"].count()
school_read_pass_per = school_read_pass/ school_students * 100


school_summary = pd.DataFrame({"School Type": school_type,
                              "Total Students": school_students,
                              "Total School Budget":school_budget.map("${:,.2f}".format),
                              "Per Student Budget": student_budget.map("${:,.2f}".format),
                              "Average Maths Score": school_avg_math,
                              "Average Reading Score": school_avg_read,
                              "% Passing Maths": school_math_pass_per,
                              "% Passing Reading": school_read_pass_per,
                              })
                            
                            
school_summary

```
